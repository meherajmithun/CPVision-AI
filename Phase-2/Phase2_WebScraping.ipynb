```python
# Install BeautifulSoup
# BeautifulSoup is a Python library that helps us read HTML.
from bs4 import BeautifulSoup
import requests

# Imagine HTML as a huge book.
# BeautifulSoup is like the table of contents that lets us quickly find the section we need.
```

```python
contest_id = 2060
index = "A"

url = f"https://codeforces.com/problemset/problem/{contest_id}/{index}"

print(url)

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

print(response.status_code)
```
