The goal of Phase 1 is to collect structured information about Codeforces problems and store it in a dataset that will later be used for machine learning.

At this stage, we are not training any AI model. We are only collecting data.

requests allows our program to communicate with the Codeforces server through its API.
pandas helps organize the collected data into a table (DataFrame), making it easy to analyze and save as a CSV file.

```python

import requests
import pandas as pd
```

```python
# Connect to the Codeforces API
url = "https://codeforces.com/api/problemset.problems"

response = requests.get(url)

# Check Whether the Request Was Successful
data = response.json()

print(data["status"])
# Sometimes

# internet may fail
# server may be down
# API may return an error

# Before using the data, we verify that the request succeeded.

# Expected output:

# OK
```

```python
print(data.keys())
print(data["result"].keys())
print(data["result"]["problems"][0])
```

Inspect One Problem

We print one problem.

```python

print(data["result"]["problems"][0])
```

```python
# We create a Python list. //problems = []

problems = []
for p in data["result"]["problems"]:

    contest_id = p.get("contestId")

    index = p.get("index")

    name = p.get("name")

    rating = p.get("rating")

    tags = ", ".join(p.get("tags", []))

    problems.append({
        "contestId": contest_id,
        "index": index,
        "name": name,
        "rating": rating,
        "tags": tags
    })
```

```python
# Convert to a DataFrame
df = pd.DataFrame(problems) #The collected records are converted into a DataFrame to facilitate data analysis and preprocessing.
df.head() #Verify the Dataset #We check whether
# columns are correct
# values are correct
# nothing is missing
```

```python
print(df.shape)
```

```python
df.to_csv("metadata.csv", index=False) #Save the Dataset
# Machine learning models require datasets stored on disk.
```
